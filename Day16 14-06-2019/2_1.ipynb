```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# load an color image
img = cv2.imread('truck.tiff',0)
(row,col)=img.shape

cv2_imshow(img)


#-------------histogram------- plot
hist=np.zeros(256)
for n in range(256):
    for i in range(row):
        for j in range(col):
            if img[i][j]==n:
                hist[n]+=1

plt.plot(hist)

plt.xlim(0, 255)
plt.xlabel('pixel values')
plt.ylabel('freq of pixels')
plt.title('histogram')
plt.show()
 
```
